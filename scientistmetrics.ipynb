## Chi-squared statistic

Let a sample of size $n$ of simultaneously distributed variables $A$ and $B$ for $i=1,\dots,r;j=1,\dots,c$ be given by the frequencies $n_{ij}$ which is the number of times the values $\left(A_{i},B_{j}\right)$ were observed. The chi - squared statistic then is :

$$
\chi^{2} = \displaystyle \sum_{i=1}^{i=r}\sum_{j=1}^{j=c} \dfrac{\left(n_{ij} - v_{ij}\right)^{2}}{v_{ij}}
$$

with $v_{ij} = \dfrac{n_{i \bullet}\times n_{\bullet j}}{n}$ where $n_{i \bullet} = \displaystyle \sum_{j=1}^{j=c}n_{ij}$ is the number of times the value $A_{i}$ is observed and $n_{\bullet j} = \displaystyle \sum_{i=1}^{i=r}n_{ij}$ is the number of times the value $B_{j}$.

```python
# chi -squared statistic
chi2 = sm.scientistmetrics(df,method="chi2")
chi2
```

```python
# Heatmap
p = sm.ggheatmap(chi2,
                 title="Chi - squared statistic heatmap",
                 tl_cex = 10,tl_srt = 90)
print(p)
```

## Phi coefficient

In [statistics](https://en.wikipedia.org/wiki/Statistics), the phi coefficient (or mean square contingency coefficient and denoted by Ï† or rÏ†) is a measure of association for two [binary variables](https://en.wikipedia.org/wiki/Binary_variables). The formula is  :

$$
\phi^{2}= \dfrac{\chi^{2}}{n}
$$

where $\chi^{2}$ is the chi - squared statistic and $n$ the total number of observations.

```python
# Phi coeffcient
phi = sm.scientistmetrics(df,method="phi")
phi
```

```python
# Heatmap
p = sm.ggheatmap(phi,
                 title="Phi coefficient association heatmap",
                 tl_cex = 10,tl_srt = 90)
print(p)
```

## Cramer's V

Cramer's V is computed by taking the square root of the chi-squared statistic divided by the sample size and the minimum dimension minus 1:

$$
V = \sqrt{\dfrac{\chi^{2}/n}{\min\left(r-1,c-1\right)}} = \sqrt{\dfrac{\phi^{2}}{\min\left(r-1,c-1\right)}}
$$


```python
# Cramer's V 
cramersv = sm.scientistmetrics(df,method="cramer")
cramersv
```

```python
# Heatmap
p = sm.ggheatmap(cramersv,
                 title="Cramer's V association heatmap",
                 tl_cex = 10,tl_srt = 90,show_diag=True)
print(p)
```

## Tschuprow's T

In statistics, Tschuprow's T is a measure of association between two nominal variables, giving a value between 0 and 1 (inclusive). It is closely related to CramÃ©r's V, coinciding with it for square contingency tables.

$$
V = \sqrt{\dfrac{\phi^{2}}{\sqrt{(r-1)(c-1)}}}
$$



```python
# Tschuprow's T
tschuprowt = sm.scientistmetrics(df,method="tschuprow")
tschuprowt
```

```python
# Heatmap
p = sm.ggheatmap(tschuprowt,
                 title="Tschuprow's T association heatmap",
                 tl_cex = 10,tl_srt = 90)
print(p)
```

## G - test

In statistics, G-tests are likelihood-ratio or maximum likelihood statistical significance tests that are increasingly being used in situations where chi-squared tests were previously recommended.

$$
G = 2\displaystyle \sum_{i}O_{i}\times \ln \left(\dfrac{O_{i}}{E_{i}}\right)
$$

where $O_{i}\geq 0$ is the observed count in a cell, $E_{i}>0$ is the expected count under the null hypothesis with :

$$
\displaystyle \sum_{i} O_{i} = \displaystyle \sum_{i} E_{i} = n
$$

where $n$ is the number of observations.

```python
# Pearson
gtest = sm.scientistmetrics(df,method="gtest")
gtest
```

```python
# Heatmap
p = sm.ggheatmap(gtest,
                 title="log-likelihood heatmap",
                 tl_cex = 10,tl_srt = 90)
print(p)
```
